```python
%sql
USE CATALOG mini_dolar_futuro;

CREATE OR REPLACE TABLE mini_dolar_futuro.gold.wdofut_dashboard AS
SELECT
    data_pregao,
    pivot_point,
    R1, 
    R2, 
    R3, 
    S1, 
    S2, 
    S3 
FROM
    mini_dolar_futuro.silver.wdofut_tratado
ORDER BY
    data_pregao DESC;

OPTIMIZE mini_dolar_futuro.gold.wdofut_dashboard
ZORDER BY (data_pregao);
```
